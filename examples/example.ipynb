```python
from pytrajlib import plot, simulation
```

```python
run_params = simulation.get_default_run_params()
run_params["num_runs"] = 100
run_params["atm_model"] = 1
run_params["impact_output"] = 1
impact_data = simulation.run(run_params)
plot.impact(run_params=run_params, impact_data=impact_data)
```
